```python
seq1 = input("first seq : ") # seq 1 입력
lseq1 = list(seq1)
seq2 = input("second seq : ") # seq 2 입력
lseq2 = list(seq2)
gap_score = input("gap score : ") # gap score 입력
match_score = input("match score : ") # match score 입력
mismatch_score = input("mismatch score : ") # mismatch score 입력
```

```python
score_matrix = [['0'] * (len(seq1)+2) for _ in range(len(seq2)+2) ]
#print(score_matrix)

#for i in range(len(seq2)+2):
#  print(score_matrix[i])

for i in range(2,len(seq1)+2,1):
  #print(i)
  score_matrix[0][i] = lseq1[i-2]
  #print(score_matrix[0][i])

for i in range(2,len(lseq2)+2,1):
  score_matrix[i][0] = lseq2[i-2]
```

```python
for i in range(1,len(seq1)+2,1):
  if i == 1:
    score_matrix[1][i] = '0'
  else:
    a = int(score_matrix[1][i-1]) + int(gap_score)
    score_matrix[1][i] = str(a)

for i in range(1,len(seq2)+2,1):
  if i == 1:
    score_matrix[i][1] = '0'
  else:
    a = int(score_matrix[i-1][1]) + int(gap_score)
    score_matrix[i][1] = str(a)
```

```python
import numpy as np
```

```python
direction_matrix = np.zeros((3,len(seq2)+2,len(seq1)+2))
```

```python
for i in range(2,len(seq2)+2,1):
  for j in range(2,len(seq1)+2,1):
    if score_matrix[i][0] == score_matrix[0][j]:
      s = match_score
    else:
      s = mismatch_score
    
    s1 = int(score_matrix[i-1][j-1]) + int(s)
    s2 = int(score_matrix[i-1][j]) + int(gap_score)
    s3 = int(score_matrix[i][j-1]) + int(gap_score)
    
    score_matrix[i][j] = str(max(s1,s2,s3))

    if str(max(s1,s2,s3)) == str(s1):
      #대각선 방향
      direction_matrix[0][i][j] = 1
    if str(max(s1,s2,s3)) == str(s2):
      direction_matrix[1][i][j] = 1
    if str(max(s1,s2,s3)) == str(s3):
      direction_matrix[2][i][j] = 1
```

```python
for i in range(len(seq2)+2):
  print(score_matrix[i])
```

```python
print(direction_matrix)
```

```python
start_seq1 = len(seq1)+1
start_seq2 = len(seq2)+1
rev_seq1 = []
rev_seq2 = []

for i in range(len(seq2)+1,1,-1):
  for j in range(len(seq1)+1,1,-1):
    print(i)
    print(j)

    if i == start_seq2 and j == start_seq1:
      print(start_seq2)
      print(start_seq1)

      if direction_matrix[0][i][j] == 1:
        sc1 = int(score_matrix[i-1][j-1])
      else:
        sc1 = -1000000
      
      if direction_matrix[1][i][j] == 1:
        sc2 = int(score_matrix[i-1][j])
      else:
        sc2 = -1000000
      
      if direction_matrix[2][i][j] == 1:
        sc3 = int(score_matrix[i][j-1])
      else:
        sc3 = -1000000

      if str(max(sc1,sc2,sc3)) == str(sc1):
        rev_seq1.append(score_matrix[0][j])
        rev_seq2.append(score_matrix[i][0])
        start_seq1 = start_seq1 - 1
        start_seq2 = start_seq2 - 1

      elif str(max(sc1,sc2,sc3)) == str(sc2):
        rev_seq1.append('-')
        rev_seq2.append(score_matrix[i][0])
        start_seq2 = start_seq2 - 1
    
      else:
        rev_seq1.append(score_matrix[0][j])
        rev_seq2.append('-')
        start_seq1 = start_seq1 - 1
      
      if start_seq2==1 and start_seq1==2:
        rev_seq1.append(score_matrix[0][start_seq1])
        rev_seq2.append('-')
        break

      elif start_seq2==2 and start_seq1==1:
        rev_seq1.append('-')
        rev_seq2.append(score_matrix[start_seq2][0])
        break
      
      print(rev_seq1)
      print(rev_seq2)
```

```python
print(rev_seq1)
print(rev_seq2)
```

```python
rev_seq1.reverse()
rev_seq2.reverse()
print(rev_seq1)
print(rev_seq2)
```

```python
best_seq1 = "".join(rev_seq1)
best_seq2 = "".join(rev_seq2)
```
